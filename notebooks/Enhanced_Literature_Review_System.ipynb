## ğŸ”¬ Enhanced Paper Analysis

## ğŸ“ Citation Tracking & Analysis

Discover precise citation locations, reference parsing, and context analysis for literature review writing.

```python
# ğŸ“ Run citation tracking analysis
print("ğŸ“ Analyzing Citations & References...")

citations = track_citations_in_paper(
    enhanced_doc['content'], 
    enhanced_doc['metadata']
)

print("\nğŸ“ CITATION ANALYSIS")
print("=" * 40)
print(f"ğŸ“„ Paper: {citations['paper_info']['title'][:60]}...")
print(f"ğŸ“Š Inline Citations: {len(citations['inline_citations'])}")
print(f"ğŸ“š Reference List: {len(citations['reference_list'])}")
print(f"ğŸ“ˆ Citation Density: {citations['citation_density']['citations_per_1000_words']:.1f} per 1000 words")

if citations['inline_citations']:
    print(f"\nğŸ” SAMPLE INLINE CITATIONS:")
    for i, cite in enumerate(citations['inline_citations'][:3]):
        print(f"  {i+1}. '{cite['text']}' (Line {cite['line_number']}, {cite['section']})")
        print(f"     Context: ...{cite['context']['sentence'][:100]}...")

if citations['reference_list']:
    print(f"\nğŸ“š SAMPLE REFERENCES:")
    for i, ref in enumerate(citations['reference_list'][:3]):
        print(f"  [{ref['number']}] {ref['authors'][:2] if ref['authors'] else 'Unknown'}")
        print(f"      Title: {ref['title'][:60]}..." if ref['title'] else "      Title: Not extracted")
        print(f"      Year: {ref['year']}" if ref['year'] else "      Year: Unknown")

if citations['key_claims']:
    print(f"\nğŸ’¡ KEY CLAIMS WITH CITATIONS:")
    for i, claim in enumerate(citations['key_claims'][:2]):
        print(f"  {i+1}. [{claim['type'].title()}] {claim['claim'][:80]}...")
        print(f"     Section: {claim['section']}")
```

## ğŸ•¸ï¸ Knowledge Graph Deep Dive

Explore entities, relationships, and discover connections in your paper.

```python
# ğŸ•¸ï¸ Knowledge Graph Overview
graph_summary = chat_system.kg.get_graph_summary()

if 'error' not in graph_summary:
    print("ğŸ•¸ï¸ KNOWLEDGE GRAPH ANALYSIS")
    print("=" * 35)
    print(f"ğŸ“Š Total Entities: {graph_summary['total_nodes']}")
    print(f"ğŸ”— Total Relationships: {graph_summary['total_edges']}")
    print(f"ğŸ¯ Graph Density: {graph_summary['graph_density']:.3f}")
    
    print("\nğŸ“ ENTITIES BY CATEGORY:")
    for category, count in graph_summary['nodes_by_category'].items():
        print(f"  â€¢ {category.title()}: {count}")
    
    print("\nâ­ MOST CONNECTED ENTITIES:")
    for item in graph_summary['most_connected'][:5]:
        print(f"  â€¢ {item['node']} (connection score: {item['connections']:.2f})")
else:
    print(f"âŒ Graph summary not available: {graph_summary['error']}")
```

```python
# ğŸ” Entity Deep Dive - Pick an interesting entity to explore
entities = chat_system.get_entities()

# Find an interesting entity
entity_to_explore = None
for category, items in entities.items():
    if items and isinstance(items, list) and len(items) > 0:
        entity_to_explore = items[0]  # Pick the first entity from any category
        break

if entity_to_explore:
    print(f"ğŸ” DEEP DIVE: '{entity_to_explore}'")
    print("=" * 50)
    
    exploration = chat_system.explore_entity(entity_to_explore)
    
    if 'error' not in exploration:
        print(f"ğŸ“Š Entity: {exploration['node']}")
        print(f"ğŸ“ Category: {exploration['category']}")
        print(f"ğŸ”— Direct Connections: {exploration['total_connections']}")
        
        if exploration['connections']:
            print("\nğŸ•¸ï¸ RELATIONSHIPS:")
            for conn in exploration['connections'][:5]:  # Show top 5
                print(f"  â€¢ {conn['relationship']} â†’ {conn['target']} ({conn['category']})")
            
            if len(exploration['connections']) > 5:
                print(f"  ... and {len(exploration['connections']) - 5} more connections")
        
        if 'rag_context' in exploration:
            print(f"\nğŸ’¬ CONTEXT FROM PAPER:")
            context = exploration['rag_context']
            print(f"  {context[:200]}{'...' if len(context) > 200 else ''}")
    else:
        print(f"âŒ Could not explore entity: {exploration['error']}")
else:
    print("âŒ No entities found to explore")
    print("ğŸ’¡ Try loading a paper with more structured content")
```
